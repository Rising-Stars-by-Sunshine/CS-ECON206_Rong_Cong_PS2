### 1 Prisoner's dilemma


**1.2 Game Description:**



The Prisoner's dilemma is a well-known concept in game theory used to describe a conflict between cooperation and competition between two people. In this game, two prisoners are taken into custody and asked if they want to plead guilty. If both plead guilty(confess), they each go to prison for 10 years; If both men plead not guilty(keep silent), they will each serve two years in prison; If one pleads guilty (confess)and the other keeps silent, the former will go to prison for 12 years and the latter will be released.

|     | keep silent | confess
|  ----  | :----:  | :----: |
| keep silent  | 2 yuears in jail, 2 yuears in jail | 12 yuears in jail, 0 yuears in jail |
| confess  | 0 yuears in jail, 12 yuears in jail |  10 yuears in jail, 10 yuears in jail |

Given that both prisoners want to spend less time in jail, n year in jail is calculated as -n for clarity.

|     | keep silent | confess
|  ----  | :----:  | :----: |
| keep silent  | -2, -2 | -12, 0 |
| confess | 0, -12 |  -10, -10 |

```python
import nashpy as nash
import numpy as np

# Creater the game with the payoff matrix

A = np.array([[-2, 0],
              [-12, -10]]) # A is the row player (in this case, prisoner)

B = np.array([[-2, -12],
              [0, -10]]) # B is the column player
              # Form the game
game2 = nash.Game(A,B)
game2

     
```

```python
# Find the Nash Equilibrium with Support Enumeration
equilibria = game2.support_enumeration()
for eq in equilibria:
    print(eq)
     
```

**1.2 Elaborate on the Nash equilibrium solution**


A strategy is a Nash Equilibrium strategy if it is the best strategy for an agent, given the strategies the other players are actually using.
The dominant strategy is best, no matter what others do, best under all possible conditions. NE strategy is best, given what others actually do. There could be a better strategy if others played strategies different from their actual strategies. The dominant strategy is best, no matter what others do, best under all possible conditions. In many games, there are no dominant strategies but there are NE strategies.

In the prisoner's dilemma, when one prisoner chooses to keep silent, the best option for the other prisoner is to confess. When one prisoner chooses to confess, the best option for another prisoner equally confesses. In other words, no matter what the other side chooses, confession is always the best choice under the circumstances.
Therefore, in this case, the NE strategy is [-10,-10],  both sides confess.

##2 Matching Pennies

**2.1 Game Description**

Matching Pennies is a two-person game. In this game, two players flip a coin at the same time, and each coin has two sides. After Player A flips A coin, Player B's goal for a player is to get him to flip a coin that is different from his opponent's coin, thus making Player A pay himself 10. Player A's goal is to get him to turn over the same coin as his opponent's coin so that player A pays himself 10. Each player has one of two strategies: choose to flip the heads or tails of the coin.

If two players turn over coins with the same side, then B gains 10 and A loses 10. If two players flip coins with different sides, then A gains 10 and B loses 10.

**2.2 Solutions**

In the game Matching Pennies, there is an extreme Nash equilibrium, which occurs when B chooses heads and A chooses tails; A chooses tails, B chooses heads.
In this case, neither player has a better strategy to win and will lose no matter which coin side they choose. Thus, in this case, both players are more likely to randomly choose their coin faces to avoid getting into an unfavorable situation.
At the same time, the expected payoff for each player depends on their strategy choices and the strategy choices of their opponents.
